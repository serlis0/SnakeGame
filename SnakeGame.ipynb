```python
%reset -f
import pygame
import time
import random
import numpy as np
pygame.init()

snake_block = 20
dis_width = 50 * snake_block
dis_height = 10 * snake_block
food_block = 8
snake_speed = 20
food_shift = (snake_block - food_block) /2

dis=pygame.display.set_mode((dis_width, dis_height))
pygame.display.set_caption('Snake game by Lysychkin')
 
green = (50, 250, 50)
red   = (255, 0, 0)
white = (255, 255, 255)
black = (0, 0, 0)
blue = (0, 0 , 255)
yellow = (200, 200, 50)

clock = pygame.time.Clock()

font_style = pygame.font.SysFont(None, 50)
score_font = pygame.font.SysFont("comicsansms", 35)
def message(msg, color):
    mesg = font_style.render(msg, True, color)
    dis.blit(mesg, [dis_width / 6, dis_height / 3])
    
def generateFood():
    food = np.array([random.randrange(0, dis_width - snake_block), 
                    random.randrange(0, dis_height - snake_block)])
    return np.around(food / snake_block) * snake_block

def Your_score(score):
    value = score_font.render("Your Score: " + str(score), True, yellow)
    dis.blit(value, [0, 0])
    
def gameLoop(): # main function    
    game_over = False
    game_close = False
    
    head = [dis_width / 2, dis_height / 2]
    snake = []
    score = 0
    
    
    x_change = 0
    y_change = 0
    
    food_x, food_y = generateFood()

    while not game_close:
        
        while game_over == True:
            dis.fill(white)
            message("You Lost! Press Q-quite or R-play!", red)
            Your_score(score)
            pygame.display.update()
            if event.type==pygame.QUIT:
                game_over = False
                game_close = True
                
            for event in pygame.event.get():
                if event.type == pygame.KEYDOWN:
                    if event.key == pygame.K_q:
                        game_over = False
                        game_close = True
                    if event.key == pygame.K_r:
                        gameLoop()
 

        for event in pygame.event.get():
            if event.type==pygame.QUIT:
                game_close = True
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_LEFT:
                    x_change = -snake_block
                    y_change = 0
                elif event.key == pygame.K_RIGHT:
                    x_change = snake_block
                    y_change = 0
                elif event.key == pygame.K_UP:
                    y_change = -snake_block
                    x_change = 0
                elif event.key == pygame.K_DOWN:
                    y_change = snake_block
                    x_change = 0


        head[0] += x_change
        head[1] += y_change
        head[0] = (head[0] + dis_width) % dis_width
        head[1] = (head[1] + dis_height) % dis_height
        dis.fill(white)
        
        for block in snake[:-1]:
            if block == head:
                game_over = True
        
        snake.append(head.copy())
        if len(snake) > score + 1:
            del snake[0]
            
        Your_score(score)
        for block in snake:
            pygame.draw.rect(dis,green,[block[0], block[1], snake_block, snake_block])
        pygame.draw.rect(dis, blue, [food_x + food_shift, food_y + food_shift,
                                     food_block, food_block])
                       
        pygame.display.update()
        
        if head[0] == food_x and head[1] == food_y:
            print("Yummy!!")
            print(food_x + food_shift, food_y + food_shift)
            food_x, food_y = generateFood()
            score += 1
        clock.tick(snake_speed)


    message("You lost!", red)
    pygame.display.update()
    time.sleep(1.2)

    pygame.quit()
    
gameLoop()
```
